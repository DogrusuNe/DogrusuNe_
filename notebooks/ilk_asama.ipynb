```python
import pandas as pd
dataf = pd.read_csv('wiki.tr.txt', header=None)
data = dataf.copy()
```

```python
data.head(20)
```

```python
with open("Sözlüklerimiz/kelime_sözlügü.txt","r",encoding="utf-8") as f:
    kelimeler = [line.strip() for line in f]
    
len(kelimeler)
```

```python
with open("Sözlüklerimiz/ayrıyazılan.txt","r",encoding="utf-8") as f:
    ayrikelimeler = [line.strip() for line in f]
    
ayrikelimeler
```

```python
with open("Sözlüklerimiz/birlesikyazılankelimeler.txt","r",encoding="utf-8") as f:
    birlesikkelimeler = [line.strip() for line in f]
    
birlesikkelimeler
```

```python
with open("Sözlüklerimiz/birlesikyazılanişaretlenmiş.txt","r",encoding="utf-8") as f:
    birlesikkelimeler2 = [line.strip() for line in f]
    
birlesikkelimeler2
```

```python
with open("Sözlüklerimiz/ayrıyazılanisaretlenmis.txt","r",encoding="utf-8") as f:
    ayrikelimeler2 = [line.strip() for line in f]
    
ayrikelimeler2
```

```python
cumleler = data[data[0].str.contains('|'.join(kelimeler))]

bitisik_cumleler = cumleler[~cumleler[0].str.contains('|'.join(birlesikkelimeler))]
ayri_cumleler = cumleler[cumleler[0].str.contains('|'.join(ayrikelimeler))]

print(bitisik_cumleler.shape)
print(ayri_cumleler.shape)
```

```python
for k in range(len(birlesikkelimeler)):
    bitisik_cumleler[0]= bitisik_cumleler[0].str.replace(birlesikkelimeler[k], birlesikkelimeler2[k], case = False) 

for k in range(len(ayrikelimeler)):
    ayri_cumleler[0]= ayri_cumleler[0].str.replace(ayrikelimeler[k], ayrikelimeler2[k], case = False) 

bitisik_cumleler = bitisik_cumleler[bitisik_cumleler[0].str.split().str.len().lt(10)]
ayri_cumleler = ayri_cumleler[ayri_cumleler[0].str.split().str.len().lt(10)]
```

```python
bitisik_cumleler = bitisik_cumleler.sample(frac=1).reset_index(drop=True)
ayri_cumleler = ayri_cumleler.sample(frac=1).reset_index(drop=True)
```

```python
bitisik_cumleler = bitisik_cumleler.sample(ayri_cumleler.shape[0]).reset_index(drop=True)
```

```python
print(ayri_cumleler.shape)
print(bitisik_cumleler.shape)
```

```python
data_bitisik = bitisik_cumleler
data_bitisik = data_bitisik.sample(frac=1).reset_index(drop=True)
data_bitisik.head(n=10)
```

```python
data_ayri = ayri_cumleler
data_ayri = data_ayri.sample(frac=1).reset_index(drop=True)
data_ayri.head(n=10)
```

```python
data_bitisik['y'] = 1
data_ayri['y'] = 0
```

```python
data_final = pd.concat([data_bitisik, data_ayri])
data_final = data_final.sample(frac=1).reset_index(drop=True)
data_final.head(n=20)
```

```python
data_final = data_final.rename(columns={0: 'line'})
```

```python
print(data_final.shape)
```

```python
data = data_final.sample(frac=1).reset_index(drop=True)
data['line'] = data['line'].str.lower() 
print(data.shape)
data.head(n=10)
```

```python
data.to_csv('ilk_asama_data.csv', encoding='utf-8', index=False)
```
